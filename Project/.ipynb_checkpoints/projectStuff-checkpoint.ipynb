# Data Processing
### Load Libraries

```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score, RepeatedKFold
import matplotlib.pyplot as plt
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler, MaxAbsScaler, MinMaxScaler, LabelEncoder
import json
import lightgbm as lgbm
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import seaborn as sns
from imblearn.over_sampling import RandomOverSampler
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import mean_squared_error, confusion_matrix, accuracy_score
from mlxtend.plotting import plot_confusion_matrix
```

### Load Data

```python
df = pd.read_csv("T_ONTIME_REPORTING.csv")
df.head()
```

```python
print(df.shape)
```

```python
df.OP_UNIQUE_CARRIER.unique()
```

### Remove the indicator columns "DELAY_INDICATOR" and "CANCELLED" as those are counted for in the INDICATOR column. Also using LabelEncoder to encode the the variables that are strings (Airport Code, Airport Codes, etc.). Since the year and month are the same for all the records, those columns will be dropped as well.

```python
df.drop(columns=["DELAY_INDICATOR","CANCELLED","YEAR", "MONTH"],inplace=True)
le = preprocessing.LabelEncoder()
df['OP_UNIQUE_CARRIER'] = le.fit_transform(df['OP_UNIQUE_CARRIER'])
df['ORIGIN'] = le.fit_transform(df['ORIGIN'])
df['DEST'] = le.fit_transform(df['DEST'])
```

### Checking for null and NaN values.

```python
df.isnull().sum()
```

Since all the NaN/null values come from flights that have been cancelled (no depart time information). Filling those null values with 0's.

```python
for index, row in df.iterrows():
    if row['INDICATOR']==2:
        df.at[index,'DEP_DELAY_NEW']=0
```

Checking to make sure that all the NaN/null values are filled.

```python
df.isnull().values.any()
```

```python
df.isna().values.any()
```

#### Even though there is no missing data, I will still "drop" rows that are missing the indicator (for good practice).

```python
df = df[df['INDICATOR']>=0]
```

```python
df.INDICATOR.value_counts()
```

61.07% of the flights were on-time  
32.75% of the flights were delayed  
6.18% of the flights were cancelled

```python
plt.pie(df.INDICATOR.value_counts(), startangle=90, autopct='%1.2f%%', labels=['on-time','delayed','cancelled'])
plt.title("Class Distribution")
plt.show()
```

```python
df.OP_UNIQUE_CARRIER.value_counts()
```

```python
df.OP_UNIQUE_CARRIER.value_counts().plot(title="Distribution of Airline Records",kind='barh',xlabel="Airline Code")
```

```python
def countX(lst, x):
    count = 0
    for ele in lst:
        if (ele == x):
            count = count + 1
    return count

cols = df.OP_UNIQUE_CARRIER.unique()

for col in cols:
    testList = df[df['OP_UNIQUE_CARRIER']==col]
    ot = countX(testList.INDICATOR, 0)
    delay = countX(testList.INDICATOR, 1)
    canc = countX(testList.INDICATOR, 2)
    piedata = [ot,delay,canc]
    print("Airline %s:\n\tOn-Time: %d\n\tDelayed: %d\n\tCancelled: %d\n\t"%(col, ot,delay,canc))
    plt.figure(figsize=(5,5))
    plt.pie(piedata, startangle=90, autopct='%1.2f%%', labels=['on-time','delayed','cancelled'])
    plt.title("Class Distribution for Airline %s" % col)
    plt.show()
```

Because the data is imbalanced, I will be using RandomOverSampler to create an even number of each class. I will not be applying this to the distribution of airline records at this moment, but with the indicator column (number of on-time, cancelled, and delayed data). Before I do that, I will need to separate the data.  
Assigning the target column INDICATOR values to Y  
Assigning the rest of the dataframe values to X

```python
Y = df.INDICATOR.values
X = df.drop(columns='INDICATOR',inplace=False).copy()
Y.shape
```

## Applying Different ML Classification methods, scaling using StandardScaler()

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y, test_size=0.25, random_state=42)

sc = StandardScaler()
sc.fit(x_train)
x_train = sc.transform(x_train)
x_test = sc.transform(x_test)
```

### Basline Model, default logistic regression (with max_iter being increased to allow the method to converge)

```python
%%time
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression(max_iter=10000)
clf.fit(x_train,y_train)
print(clf.score(x_test,y_test))
from sklearn.metrics import ConfusionMatrixDisplay
cm = ConfusionMatrixDisplay.from_estimator(clf, x_test, y_test)
#plt.figure()
#plot_confusion_matrix(cm, figsize=(12,8), hide_ticks=True, cmap=plt.cm.Blues)
plt.title("Logistic Regression Model Default - Confusion Matrix")
plt.xticks(range(3), ["Class 0","Class 1","Class 2"], fontsize=16)
plt.yticks(range(3), ["Class 0","Class 1","Class 2"], fontsize=16)
plt.show()
scores_log=cross_val_score(clf,x_train,y_train,cv=5)
print(scores_log)
```

### Logistic Regression with solver 'sag'
#### Cross validation using cv=5 

```python
%%time
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression(max_iter=10000, solver='sag',n_jobs=-1)
clf.fit(x_train,y_train)
print(clf.score(x_test,y_test))
from sklearn.metrics import ConfusionMatrixDisplay
cm = ConfusionMatrixDisplay.from_estimator(clf, x_test, y_test)
#plt.figure()
#plot_confusion_matrix(cm, figsize=(12,8), hide_ticks=True, cmap=plt.cm.Blues)
plt.title("Logistic Regression Model 'SAG' - Confusion Matrix")
plt.xticks(range(3), ["Class 0","Class 1","Class 2"], fontsize=16)
plt.yticks(range(3), ["Class 0","Class 1","Class 2"], fontsize=16)
plt.show()
scores_log=cross_val_score(clf,x_train,y_train,cv=5)
print(scores_log)
#print("Mean score: %0.2f",scores_log.mean())
```

### Logistic Regression with solver 'saga'
#### Cross validation using cv=5 

```python
%%time
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import ConfusionMatrixDisplay
clf=LogisticRegression(max_iter=10000, solver='saga', n_jobs=-1)
clf.fit(x_train,y_train)
print(clf.score(x_test,y_test))
cm = ConfusionMatrixDisplay.from_estimator(clf, x_test, y_test)
#plt.figure()
#plot_confusion_matrix(cm, figsize=(12,8), hide_ticks=True, cmap=plt.cm.Blues)
plt.title("Logistic Regression Model 'SAGA' - Confusion Matrix")
plt.xticks(range(3), ["Class 0","Class 1","Class 2"], fontsize=16)
plt.yticks(range(3), ["Class 0","Class 1","Class 2"], fontsize=16)
plt.show()
scores_log=cross_val_score(clf,x_train,y_train,cv=5)
print(scores_log)
```

### KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
scores = []
for k in range(1,30):
    knn = KNeighborsClassifier(n_neighbors=k, n_jobs=-1)
    knn.fit(x_train,y_train)
    y_pred=knn.predict(x_test)
    #print("K: ",k,"Train Score: ",knn.score(x_train,y_train), "Test Score: ",knn.score(x_test,y_test))
    scores.append(knn.score(x_test,y_test))
print("Best K Score: ",max(scores))
print("Best K Value: ",scores.index(max(scores))+1)
```

### KNN Classifier using best K value from above

```python
knn_best = KNeighborsClassifier(n_neighbors=scores.index(max(scores))+1)
knn_best.fit(x_train,y_train)
knn_best.score(x_test,y_test)
cm = ConfusionMatrixDisplay.from_estimator(knn_best, x_test, y_test)
#plt.figure()
#plot_confusion_matrix(cm, figsize=(12,8), hide_ticks=True, cmap=plt.cm.Blues)
plt.title("Best KNN Classifier Model - Confusion Matrix")
plt.xticks(range(3), ["Class 0","Class 1","Class 2"], fontsize=16)
plt.yticks(range(3), ["Class 0","Class 1","Class 2"], fontsize=16)
plt.show()
```

```python
knn_best.score(x_test,y_test)
```

#### Cross validation using cv=5

```python
k_scores = []
for k in range(1,30):
    knn = KNeighborsClassifier(n_neighbors=k,n_jobs=-1)
    scores = cross_val_score(knn, x_train, y_train, cv=5, scoring='accuracy')
    k_scores.append(scores.mean())
print(k_scores)
```

## More advanced ML Classification Models
### Starting with XGBClassifier

```python
import xgboost as xgb
learning_rates = [0.05, 0.1, 0.25, 0.5, 0.75, 1]
xgbscores = []
for lr in learning_rates:
    xgb_clf = xgb.XGBClassifier(max_depth=10, n_estimators=100, learning_rate=lr)
    xgb_clf.fit(x_train, y_train)
    y_pred = xgb_clf.predict(x_test)
    xgbscores.append(xgb_clf.score(x_test,y_test))
    print(xgb_clf.score(x_test,y_test))
print("Best learning rate is ",learning_rates[xgbscores.index(max(xgbscores))]," with a score of ",max(xgbscores))
```

### Light GBM 
The reason for choosing this model is because it has a faster training speed and has better accuracy than other boosting algorithms, but MAINLY because it is able to handle large datasets with relative ease.

```python
import lightgbm as lgbm

learning_rates = [0.05, 0.1, 0.25, 0.5, 0.75, 1]
lgbmscores = []
for lr in learning_rates:
    lgbm_clf=lgbm.LGBMClassifier(num_leaves=100,n_estimators=100,max_depth=10,learning_rate=lr)
    lgbm_clf.fit(x_train,y_train)
    print(lgbm_clf.score(x_test,y_test))
    lgbmscores.append(lgbm_clf.score(x_test,y_test))
print("Best learning rate is ",learning_rates[lgbmscores.index(max(lgbmscores))]," with a score of ",max(lgbmscores))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, lgbm_clf.predict(x_test)))
```

```python
from sklearn.ensemble import HistGradientBoostingClassifier
learning_rates = [0.05, 0.1, 0.25, 0.5, 0.75, 1]
histscores = []
for lr in learning_rates:
    hist_clf=HistGradientBoostingClassifier(learning_rate=lr)
    hist_clf.fit(x_train,y_train)
    print(hist_clf.score(x_test,y_test))
    histscores.append(hist_clf.score(x_test,y_test))
print("Best learning rate is ",learning_rates[histscores.index(max(histscores))]," with a score of ",max(histscores))
```

## Finetuning Light GBM
Changing learning rates, bagging_fraction, bagging frequency, and the number of leaves.  
There's going to be a lot of output, so I will also be storing the output to a file to read from.

```python
import lightgbm as lgbm

outputFile = open("LGBM_output.txt","w")

learning_rates = [0.05, 0.1, 0.25, 0.5, 0.75, 1]
bagging_fractions = [.1,.2,.3,.4,.5,.6,.7,.8,.9]
bagging_frequency = [5,10,15,20]
numleaves = [10,25,50,100]
lgbmscores = []

for lr in learning_rates:
    for bfrac in bagging_fractions:
        for bfre in bagging_frequency:
            for leaves in numleaves:
                lgbm_clf=lgbm.LGBMClassifier(num_leaves=leaves,n_estimators=100,max_depth=10,learning_rate=lr, bagging_fraction=bfrac, bagging_freq=bfre)
                lgbm_clf.fit(x_train,y_train)
                lgbmscores.append(lgbm_clf.score(x_test,y_test))
                #print("Frequency: ",bfre,"\tFraction: ",bfrac,"\tLearning Rate: ",lr,"\tScore: ",lgbm_clf.score(x_test,y_test))
                outputString = "Frequency: "+str(bfre)+"\tFraction: "+str(bfrac)+"\tLearning Rate: "+str(lr)+"\tLeaves: "+str(leaves)+"\tScore: "+str(lgbm_clf.score(x_test,y_test))
                outputFile.write(outputString)
                outputFile.write('\n')
#print("Best learning rate is ",learning_rates[lgbmscores.index(max(lgbmscores))]," with a score of ",max(lgbmscores))
outputFile.close()
```
